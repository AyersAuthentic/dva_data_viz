```python
def get_month_year_us(date):
    date_list = date.split("-")
    year = date_list[0]
    month = date_list[1].lstrip("0")
    month_year_list = [month, year]
    month_year = "/".join(month_year_list)
    return month_year
```

```python
us_states['month_year'] = us_states['date'].apply(lambda date: get_month_year_us(date))
```

```python
us_states.head()
```

```python
def get_state_fips_us(geoid):
    geoid_list = geoid.split("-")
    state_fips = geoid_list[1]
    return state_fips
```

```python
us_states['state_fips'] = us_states['geoid'].apply(lambda geoid: get_state_fips_us(geoid))
```

```python
us_states.head()
```
