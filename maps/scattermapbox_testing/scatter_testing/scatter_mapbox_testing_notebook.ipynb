# Adding retrieveing fips from geo id

```python
def get_fips(geoid):
    
    id_list = geoid.split("-")
    fips = id_list[1]
    return fips
```

```python
df_covid['fips'] = df_covid['geoid'].apply(lambda x: int(get_fips(x)))
```

```python
lat_dict = pd.Series(df_loc.clat10.values, index= df_loc.fips).to_dict()
```

```python
lat_dict[36998] = 40.7128
```

# Adding lat and long columns for center of county

```python
def get_lon(fips):
    try:
        lon = lon_dict[fips]
    except:
        lon = np.NaN
    
    return lon
```

```python
def get_lat(fips):
    try:
        lat = lat_dict[fips]
    except:
        lat = np.NaN
    return lat
```
