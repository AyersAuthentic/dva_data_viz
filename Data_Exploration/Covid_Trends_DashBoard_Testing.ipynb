```python
lockdowns = pd.read_csv("U.S._State_and_Territorial_Stay-At-Home_Orders__March_15__2020___August_15__2021_by_County_by_Day.csv")
```

```python
lockdowns.head()
```

```python
lockdowns.tail()
```

```python
lockdowns.Stay_at_Home_Order_Recommendation.unique()
```

```python
lockdowns["Stay_at_Home_Order_Recommendation"] = lockdowns["Stay_at_Home_Order_Recommendation"].fillna(0)
```

```python
lockdowns.head()
```

```python
lockdowns.Stay_at_Home_Order_Recommendation.unique()
```

```python
lock_dict = {0:0,'No order for individuals to stay home': 0, 'Advisory/Recommendation': 1, 
             'Mandatory only for at-risk individuals in certain areas of the jurisdiction': 2,
            'Mandatory only for at-risk individuals in the jurisdiction': 3,
            'Mandatory only for all individuals in certain areas of the jurisdiction': 4,
            'Mandatory for all individuals': 5}
```

```python
def get_lock_level(lock_code):
    
    return lock_dict[lock_code]
```

```python
lockdowns['lockdown'] = lockdowns['Stay_at_Home_Order_Recommendation'].apply(lambda x: get_lock_level(x))
```

```python
lockdowns.tail()
```

```python
lockdowns_clean = lockdowns[['State_Tribe_Territory', 'County_Name', 'FIPS_State', 'FIPS_County', 'date', 'lockdown']]
```

```python
lockdowns_clean.head()
```

```python
lockdowns_clean.to_csv("lockdowns_data.csv")
```

```python
def pad_state(fips_state):
    state_str = str(fips_state)
    state_fips = state_str.rjust(2,'0')
    return state_fips
```

```python
lockdowns_clean['state_fips'] = lockdowns_clean['FIPS_State'].apply(lambda x: pad_state(x))
```

```python
lockdowns_clean
```

```python
def pad_county(fips_county):
    county_str = str(fips_county)
    county_fips = county_str.rjust(3, '0')
    return county_fips
```

```python
lockdowns_clean['county_fips'] = lockdowns_clean['FIPS_County'].apply(lambda x: pad_county(x))
```

```python
lockdowns_clean.head()
```

```python
lockdowns_clean['fips'] = lockdowns_clean['state_fips'] + lockdowns_clean['county_fips']
```

```python
lockdowns_clean.head()
```

```python
lock_clean = lockdowns_clean[['State_Tribe_Territory', 'County_Name', 'date', 'lockdown','fips' ]]
```

```python
lock_clean.head()
```

```python
lock_clean.to_csv("lock_clean.csv")
```

```python
def get_year_month(ymfips):
    split_st = ymfips.split("-")
    year = split_st[0]
    month_dirty = split_st[1]
    month_clean = month_dirty.rjust(2, '0')
    clean_list = [month_clean, year]
    month_year = "/".join(clean_list)
    return month_year
    
```

```python
df_grouped['month_year'] = df_grouped['YMFIPS_x'].apply(lambda x: get_year_month(x))
```
