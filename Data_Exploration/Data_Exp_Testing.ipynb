```python
states = pd.read_csv("us-states.csv")
```

```python
states.head()
```

```python
states['temp'] = states['date'] +"-"+ states['geoid']
```

```python
states['temp'].head()
```

```python
def get_YMFIPS(temp):
    
    splitted = temp.split("-")
    splitted.pop(2)
    splitted.pop(2)
    YMFIPS = "-".join(splitted)
    return YMFIPS
```

```python
states["YMFIPS"] = states['temp'].apply(lambda x: get_YMFIPS(x))
```

```python
states.head()
```

```python
states_focused = states[['cases_avg_per_100k', 'YMFIPS']].groupby(["YMFIPS"]).sum()
```

```python
states_focused
```

```python
states_focused.reset_index(inplace=True)
```

```python
states_focused
```

```python
def clean_ymfips(ymfips):
    listy = ymfips.split("-")
    month = listy[1]
    month = month.lstrip("0")
    listy[1] = month
    clean_ymfips = "-".join(listy)
    return clean_ymfips
```

```python
states_focused['YMFIPS_c'] = states_focused['YMFIPS'].apply(lambda x: clean_ymfips(x))
```

```python
states_focused.head()
```

```python
states_focused
```

```python
main_c = main.merge(states_focused, how='left', left_on = 'YMFIPS', right_on = 'YMFIPS_c')
```

```python
main_c.head()
```

```python
main_c['cases_avg_per_100k'].isna().sum()
```
